# Runnables in Langchain


```python
import random

class NakliLLM:
    def __init__(self):
        print("LLM Created!")

    def predict(self,prompt):
        response_list = [
            "Delhi is capital of India.",
            "RCB is best team in IPL",
            "AI stands for Artificial Intelligence"
        ]

        print("Here is your Response -> ")
        return {'response': random.choice(response_list)}
```

```python
llm = NakliLLM()
llm.predict("what is capital of India")
```

```python
class NakliPromptTemplate:

    def __init__(self,template,input_variable):
        self.template = template
        self.input_variable = input_variable
    
    def format(self,input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template="Write a poem on {topic}",
    input_variable=['topic']
)

template.format({'topic':'India'})
```

## Now we can use these custom components to build LLM apps

```python
prompt = template.format({'topic':'RCB'})
llm.predict(prompt)
```

### Let's build a chain class to connect these two components

```python
class NakliLLMChain:
    
    def __init__(self,template,llm):
        self.llm = llm
        self.template=template

    def run(self,input_dict):
        final_prompt = self.template.format(input_dict)
        print(final_prompt)
        result = self.llm.predict(final_prompt)
        return result['response']

```

```python
LLMChain = NakliLLMChain(template,llm)
LLMChain.run({'topic':'Cricket'})
```
