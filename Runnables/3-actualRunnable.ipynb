```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.3-70b-versatile")
```

## Types of Runnables in langchain

- Task Specific Runnables : core langchain components that have been converted to runnables so they can be used in pipelines.
* eg. ChatOpenAi, PromptTemplate, Retriever etc..

- Runnables Primitives : These are fundamental building blocks for structuring execution logic in AI workflows.
* Help orchestrate execution by defining how different Runnables interact.
* eg. RunnableSequence, RunnableParallel, RunnableMap, Runnablebranch, RunnableLambda, RunnablePassthrough

## 1. RunnableSequence

- sequential chain of runnables in Langchain that executes each step one after another, passing the output of one step as input to next.

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence
```

```python
prompt1 = PromptTemplate(
    template='Write a joke about {topic}.',
    input_variables=['topic']
)

parser = StrOutputParser()

chain1 = RunnableSequence(prompt1,model,parser)
```

```python
prompt2 = PromptTemplate(
    template='Explain the joke \n {response}',
    input_variables=['response']
)

chain2 = RunnableSequence(prompt2,model,parser)

final_chain = RunnableSequence(chain1,chain2)

final_chain.invoke({'topic':'Cricket'})
```

## 2. RunnableParallel

- allows multiple runnables to execute in parallel
- each runnable receives same input and processes it independently, producing a dictionary of output

```python
from langchain_core.runnables import RunnableParallel
```

```python
prompt3 = PromptTemplate(
    template='Generate a Tweet about {topic}.',
    input_variables=['topic']
)
prompt4 = PromptTemplate(
    template='Generate a Linkedin Post about {topic}.',
    input_variables=['topic']
)

parallel_chain = RunnableParallel(
    {
        "tweet": RunnableSequence(prompt3,model,parser),
        "linkedin" : RunnableSequence(prompt4,model,parser)
    }
)

result=parallel_chain.invoke({'topic':'AI'})
print(result)
```
