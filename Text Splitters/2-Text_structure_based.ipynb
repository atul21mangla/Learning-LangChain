## Recursive Text Splitter

- \n\n - paragraph
- \n - lines/sentences
- ' ' - words
- '' - characters





### Example

**Input Text:**
```
My name is Atul Mangla. (23)
I am an AI Engineer.  (20)

I live in Haryana.  (18)
How are you doing?  (18)
```

Suppose the chunk size is **10**.

- The splitter first tries to break by paragraphs, but each paragraph is >10 chars.
- It then tries to break by lines/sentences.
- If a line is still >10 chars, it breaks by words.  

(Adjust the chunking logic as needed for your actual implementation.)

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
My name is Atul Mangla
I am an AI Engineer

I live in Haryana.
How are you doing?
"""

splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n"," ", ""],
    chunk_size=10,
    chunk_overlap=0,
)

result = splitter.split_text(text)
for i, chunk in enumerate(result):
    print(f"Chunk {i}: '{chunk}' -> Length: {len(chunk)}")
````
