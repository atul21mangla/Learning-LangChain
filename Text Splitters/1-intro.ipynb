## Text Splitting

Text splitting is process of breaking large chunks of text into smaller pieces (chunks) that an LLM can handle efficiently.

### Need for Text Splitting

- Overcoming model Limitations : Context window limit of LLMs

- Downstream Tasks - text splitting improves nearly every LLM powered task

</br>

|Task|Why spliiting helps|
|---|---|
|Embedding|Short chunks yield more accurate vectors|
|Semantic Search|Search results point to focused info, not noise|
|Summarization|Prevents hallucination and topic drift|

## Length Based

Types of length-based splitting:

- Token-based: Splits text based on the number of tokens, which is useful when working with language models.
- Character-based: Splits text based on the number of characters, which can be more consistent across different types of text.

````python
from langchain_text_splitters import CharacterTextSplitter

text = """
This is a sample text for splitting.
It contains multiple paragraphs and sentences to demonstrate how text splitters work.
Text splitting is essential for processing large documents with language models.
"""

splitter = CharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=0,
    separator=''
)

result = splitter.split_text(text)

print(result)
````

You can choose between:
> .split_text — returns plain string chunks.

> .create_documents — returns LangChain Document objects, useful when metadata needs to be preserved for downstream tasks.

> .split_documents - receives Document Objects and splits them
