```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

# Pydantic

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title:str=Field(description="Title of Movie.")
    year:int=Field(description="The year the movie was released.")
    director:str=Field(description="The director of movie.")
    rating:float=Field(description="The movie rating out of 10.")

model_with_structed_output=model.with_structured_output(Movie)
```

```python
res1 = model.invoke("Provide the details about the movie Inception")
print(res1)
```

```python
res2 = model_with_structed_output.invoke("Provide the details about the movie Inception")
print(res2)
```

### Message output alongside parsed Structure

```python
model_with_structed_output2=model.with_structured_output(Movie, include_raw=True)
model_with_structed_output2.invoke("Provide the details about the movie Inception")
```
