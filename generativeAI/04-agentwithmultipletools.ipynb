```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="qwen/qwen3-32b")
```

## Tavily Search Tool

```python
from langchain_tavily import TavilySearch

tavily_search_tool = TavilySearch(
    max_results=5,
    topic = "general"
)

tavily_search_tool.invoke("What's latest AI news in market?")
```

## Calculator Tool

```python
from langchain.tools import tool
@tool("calculator", description="performs arithmetic calculations. Use this for any math problem.")
def calc(exp:str)-> str:
    """Evaluate mathematical expression."""
    return str(eval(exp))
```

```python
from langchain.agents import create_agent

agent = create_agent(model=model,
tools = [tavily_search_tool,calc],
system_prompt="You are a helpful assistant who provides every news answer with proper citations and sources mentioned."
)
```

```python
user_input = "What's current Ai news for Anthropic and then calculate 5 + 5."

for step in agent.stream(
    {"messages":user_input},
    stream_mode="values"
):
    step["messages"][-1].pretty_print()
```
