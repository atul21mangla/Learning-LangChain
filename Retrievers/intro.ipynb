## Retrievers

A retriever is a component in Langchain that fetches relevant documents from a data source in response to a user's query.

There are multiple types of retrievers in Langchain.

All retrievers in Langchain are Runnables.




### Types of Retrievers

1. Data Source Based - Wikipedia Retriever, Vector Store Based retrievers, etc..

2. Search Strategy Based - Similarity Search, Max Marginal Relevance Search, Contextual compression etc..

---

### Wikipedia Retriever

that queries the wikipedia API to fetch relevant content for given query.

How it works:

1. You give it a query.
2. It sends the query to the wikipedia API and fetches relevant content.
3. It returns the Langchain Document object.

```python
from langchain_community.retrievers import WikipediaRetriever
retriever = WikipediaRetriever(top_k_results=2, lang='en')

query = "Winner of IPL 2025?"

docs = retriever.invoke(query)

print(docs)
```

```python
for i, doc in enumerate(docs):
    print(f"Document {i+1}:")
    print(f"Title: {doc.metadata['title']}")
    print(f"Content: {doc.page_content}")  # Print first 200 characters
    print("\n---\n")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
documents=[
    Document(page_content="Langchain helps developer build LLM applications easiy."),
    Document(page_content="Langchain provides retrievers to fetch relevant documents for a query."),
    Document(page_content="Langchain provides vector store retriever to fetch relevant documents from a vector store."),
    Document(page_content="Chroma is a vector store that can be used with Langchain's vector store retriever."),
    Document(page_content="OpenAI provides embedding models that can be used to convert documents into vectors for vector store retriever.")
]
```

```python
from langchain_nvidia_ai_endpoints import NVIDIAEmbeddings

embeddings = NVIDIAEmbeddings(model="nvidia/llama-nemotron-embed-vl-1b-v2")
```

```python
vectorstore = Chroma.from_documents(documents, embeddings,collection_name="my-documents")
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})
```

```python
query = "What is Chroma used for?"
docs = retriever.invoke(query)

for i,doc in enumerate(docs):
    print(f"----- Result {i+1} ------")
    print(doc.page_content)
```

> This type of retrieving can also be done using - vectorstore.similarity_search(query,k) but our retriever runnable is more efficient when we need to search by some other search strategy not only similarity search.
