Improves Retrieval quality by compressing documents after retrieval - keeping only relevant content based on the user's query.

> Query: "What is photosynthesis?"

**Retrieved Document by Traditional Retriever:**
"Photosynthesis is the process by which green plants and some other organisms use sunlight to synthesize foods. The grand canyon is a famous natural site.Many tourists visit it every year."

❌ Problem: The retrieved document contains irrelevant information about the grand canyon which can confuse the downstream LLM.

Only one sentence is actually relevant to query.

✅ What contextual compression retriever does is - It compresses the retrieved document to keep only the relevant content based on the user's query.

---

### How it works?

1. Base retriever retrieves N documents for a query.
2. A compressor (LLM) is applied to each document.
3. The compressor keeps only the parts relevant to query.
4. Irrelevant content is discarded.


### When to use

1. Your documents are long and contain mixed information.
2. You want to reduce context length for LLMs.
3. You need to improve answer accuracy in RAG pipelines.


### Code Example

```python
llm = ChatOpenAI(model="gpt-3.5-turbo")
compressor = LLMChainExtractor(llm)

compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor,
    base_retriever=your_base_retriever
)

query = ""

Compressed_results = compression_retriever.invoke(query)
```
