> "How can we pick results that are not only relevant to the query but also different from each other."

MMR is an information retrieval algorithm designed to reduce redundancy in the retrieved results while maintaining high relevance to the query. 

### Why MMR Retriever?

In a regular similarity search, you might get multiple documents that are very similar to each other, which can be redundant. MMR helps to diversify the results by selecting documents that are not only relevant to the query but also different from each other.

MMR avoid that by - picking the most relevant doc first, then picking the next most relevant and least similar to already selected dos, and so on until we have k documents.

```python
from langchain_core.documents import Document

documents=[
    Document(page_content="Langchain helps developer build LLM applications easiy."),
    Document(page_content="Langchain provides retrievers to fetch relevant documents for a query."),
    Document(page_content="Langchain provides vector store retriever to fetch relevant documents from a vector store."),
    Document(page_content="Chroma is a vector store that can be used with Langchain's vector store retriever."),
    Document(page_content="OpenAI provides embedding models that can be used to convert documents into vectors for vector store retriever.")
]
```

```python
from langchain_community.vectorstores import FAISS

from langchain_nvidia_ai_endpoints import NVIDIAEmbeddings

embeddings = NVIDIAEmbeddings(model="nvidia/llama-nemotron-embed-vl-1b-v2")
```

```python
vectorStore = FAISS.from_documents(documents, embeddings)

retriever = vectorStore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 2, "lambda_mult": 0.5}
    )
# close to 0 means more diverse results.
```

```python
query = "What is Langchain?"
docs = retriever.invoke(query)
```

```python
for i,doc in enumerate(docs):
    print(f"----- Result {i+1} ------")
    print(doc.page_content)
```
