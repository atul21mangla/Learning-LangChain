Retriver that lets you search and fetch relevant documents from a vector store. It uses the vector store's semantic similarity search using vector embeddings.

### How it Works:
1. You store your documents in a vector store (like FAISS, Chrma, Weaviate, etc.).
2. Eachdocument is converted into dense vector using an embedding model (like OpenAI's text-embedding-3-small, text-embedding-3-large, etc.).
3. When user enters a query:
    - It;s also tuned into vector.
    - The retriever compared the query vector with document vectors in the vector store using similarity search.
    - it returns the top k most relevant documents
