Sometimes a single query might not capture all the ways information in phrased in your deocuments.

For Example:

> "How can i stay Healthy?"

could mean - What should i eat, How much should i sleep, What exercises should i do, etc.

A simple similarity search might miss documents that talk about these things but don't use word "healthy".

1. Takes your original query.
2. Uses an LLM to generate multiple semantically different versions of that query.
3. Performs retrieval for each sub-query.
4. Combines and deduplicates the results.

Motive behind this is - How we can remove ambiguity in user's query and generate multiple different versions of the query to capture all the relevant documents.

---

## Code Example

```python
multiquery_retriever = MultiQueryRetriever.from_llm(
    llm=ChatOpenAI(model="gpt-3.5-turbo"),
    retriever=vectorStore.as_retriever(),
    query_generator_prompt=PromptTemplate.from_template(
        "Generate 3 different versions of the following query: {query}"
    ),
    # optional - how many results to return for each sub-query
    k=2
)
```
