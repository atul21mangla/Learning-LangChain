# RAG - Retrieval Augmented Generation

> Information Retrieval + Text Generation

Steps involved in RAG:
1. Indexing (process of creating external Knowledge Base)
2. Retrieval (fetching relevant documents from the knowledge base for a given query)
3. Augmentation (combining retrieved documents with the original query to create a prompt for the LLM)
4. Generation (LLM generates a response based on the augmented prompt)

---

## 1. Indexing

`Document Ingestion` : You load source knowledge into memory.(Loaders)

`Text Chunking` : You split the documents into smaller pieces (chunks) to make them easier to process and retrieve. (Text Splitters)

`Embedding Generation` : You convert the text chunks into vector representations (embeddings) that capture their semantic meaning. (Embeddings)

`Store in Vector Store` : You save the embeddings in a vector database that allows for efficient similarity search. (Vector Stores)

---

## 2. Retrieval

Now we need to find which portions of our knowledge base are relevant to a user's query. This is where retrievers come in.

We rank the documents based on their relevance to the query and return the most relevant ones.

---

## 3. Augmentation

After retrieving relevant documents, we need to combine them with the user's original query to create a prompt that the LLM can understand and generate a response from. This process is called augmentation.

---

## Why RAG?

- Private Data Access
- Up-to-date Information
- Hallucination Reduction
- Cost Reduction

## Benefits over Fine-Tuning

- No need for large labeled datasets
- No need to train again and again for new data
- Can access up-to-date information without retraining
